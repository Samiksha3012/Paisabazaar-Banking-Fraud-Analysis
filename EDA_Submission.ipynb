## 3. ***Data Wrangling***

## Data Wrangling Code

**Changing Data Types**

````python
# Write your code to make your dataset analysis ready.

# Changing Dtypes to integer
df['Age'] = df['Age'].astype('int64')
df['Num_Bank_Accounts'] = df['Num_Bank_Accounts'].astype('int64')
df['Num_Credit_Card'] = df['Num_Credit_Card'].astype('int64')
df['Num_of_Loan'] = df['Num_of_Loan'].astype('int64')
df['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].astype('int64')
df['Credit_History_Age'] = df['Credit_History_Age'].astype('int64')
df['Delay_from_due_date'] = df['Delay_from_due_date'].astype('int64')
df['Num_Credit_Inquiries'] = df['Num_Credit_Inquiries'].astype('int64')

# Changing data type to object
df['ID'] = df['ID'].astype('object')
df['Customer_ID'] = df['Customer_ID'].astype('object')
````

````python
# Checking if the data types are changed
df.info()
````

**Dropping Unnecessary columns**

````python
# Dropping columns Name and SNN as they are not useful for our analysis and we have their unique identification number.
df.drop(['Name','SSN'],axis=1,inplace=True)

# Checking if the columns are dropped
df.columns
````

**Rounding the float values for standardization and better readability**

````python
# Rounding the numbers in float cols to 2.
df['Monthly_Inhand_Salary'] = df['Monthly_Inhand_Salary'].round(2)
df['Total_EMI_per_month'] = df['Total_EMI_per_month'].round(2)
df['Amount_invested_monthly'] = df['Amount_invested_monthly'].round(2)
df['Monthly_Balance'] = df['Monthly_Balance'].round(2)
df['Credit_Utilization_Ratio'] = df['Credit_Utilization_Ratio'].round(2)

pd.set_option('display.max_columns',None)
df.head()
````
